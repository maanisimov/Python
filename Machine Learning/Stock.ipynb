```python
import numpy as np
import pandas as pd
```

```python
stock = pd.read_csv('RedditNews.csv')
```

```python
stock.head()
```

```python
stock.info()
```

```python
stock['Date'] = pd.to_datetime(stock['Date'])
stock.info()
```

```python
mask = (stock['Date'] > '2010-06-01') & (stock['Date'] <= '2016-06-01')
```

```python
stock.loc[mask].tail()
```

```python
datetimes = np.array(stock['Date'])[::100]
```

```python
datetimes[::-1]
```

```python
len(datetimes)
```

```python
mask = (stock['Date'] == datetimes[0]) # условия для первого сабсета
new = stock.loc[mask] # создание первого сабсета
new['ShiftedDate'] = new['Date'].shift(1) # создание нового столбца со смещение даты
new.head()
```

```python
for i in range(1, len(datetimes)): # пробегаемся от первой даты до последней из списка
    mask = (stock['Date'] == datetimes[i]) # создаем условие для сабсета
    temporary = stock.loc[mask] # создаем сабсет
    temporary['ShiftedDate'] = temporary['Date'].shift(1) #создаем новый столбец сабсета со смещенной датой
    new = new.append(temporary) # добавляем текущий сабсет к уже созданным
```

```python
new.head()
```

```python
new = new[new['ShiftedDate'].notnull()] # убираем строки с отсутствующими значениями ShiftedDate 
new.head()
# В случае с котировками просто уберу те фьючи, дата экспирации которых совпадает с датой торгового дня
```
