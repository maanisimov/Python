```python
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from sklearn.ensemble import  RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
```

```python
RANDOM_SEED = 100
```

```python
students['alc'] = (students['Dalc'] * 5 + students['Walc']*2) / 7 # создаём целевую переменную
students['alc'] = students['alc'].apply(lambda x: 0 if x < 3 else 1) # бинаризуем по порогу 3
students['alc'].tail()
```

```python
y = students['alc'] # вектор значений целевой функции
```

```python
X = students.drop(['Dalc', 'Walc', 'alc'], axis=1) # объясняющие переменные
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=RANDOM_SEED) 
```

```python
skf = StratifiedKFold(n_splits=4, random_state=RANDOM_SEED, shuffle=True) # задаем параметры для кросс-валидации
```

```python
metrica = make_scorer(score_func=accuracy_score) # задаем метрику качества для поиска оптимальных параметров
```

### Random Forest Classifier

```python
rfc_opt = RandomForestClassifier(n_estimators=25, criterion='gini', n_jobs=-1, random_state=RANDOM_SEED)
```

```python
rfc_opt.fit(X_train, y_train)
y_pred = rfc_opt.predict(X_test)
print('Accuracy score при использовании оптимальные параметров RFC из статьи:', accuracy_score(y_pred, y_test))
```

```python
rfc = RandomForestClassifier(random_state=RANDOM_SEED)
```

```python
rfc_param = {"n_estimators": np.arange(10, 101, 10), "max_depth": np.arange(1, 20), "criterion": ["gini", "entropy"]} 
# задали значения гиперпараметров, из которых будем выбирать наилучшие
```

```python
# подбор лучших параметров с помощью GridSearchCV
grd_rfc = GridSearchCV(estimator=rfc, param_grid=rfc_param, scoring=metrica, cv=skf, n_jobs=-1)
grd_rfc.fit(X_train, y_train)
grd_rfc.best_params_, grd_rfc.best_score_
```

```python
print('Лучшие параметры:', grd_rfc.best_params_)
print('Лучший accuracy score:', grd_rfc.best_score_)
```

```python
best_rfc = RandomForestClassifier(n_estimators=30, max_depth=10, random_state=RANDOM_SEED, n_jobs=-1)
```

```python
best_rfc.fit(X_train, y_train)
y_pred = best_rfc.predict(X_test)
```

```python
feat_imp = pd.Series(best_rfc.feature_importances_, X_train.columns).sort_values(ascending=False)[:10]
```

```python
feat_imp # таблица важности признаков
```

```python
grd_rfc.grid_scores_
```

```python
cv_res = pd.DataFrame(grd_rfc.cv_results_)
cv_res.head()
```

```python
rfc_bestres = cv_res.sort_values('mean_test_score', ascending=False)[:5]
rfc_bestres
```

```python
rfc_acc_scores = []

for i in range(5):
    rfc_5 = RandomForestClassifier(n_estimators=rfc_bestres.iloc[i]['param_n_estimators'],
                                  max_depth=rfc_bestres.iloc[i]['param_max_depth'],
                                  criterion=rfc_bestres.iloc[i]['param_criterion'],
                                  random_state=RANDOM_SEED)
    rfc_5.fit(X_train, y_train)
    y_pred = rfc_5.predict(X_test)
    rfc_acc_scores.append(accuracy_score(y_pred, y_test))
    print('n_estimators:', rfc_bestres.iloc[i]['param_n_estimators'], ',',
          'max_depth:', rfc_bestres.iloc[i]['param_max_depth'], ',',
          'criterion:', rfc_bestres.iloc[i]['param_criterion'])
    print('accuracy score:', accuracy_score(y_pred, y_test))
    print(40 * '-') 
```

```python
best_rfc_acc_score = max(rfc_acc_scores)
print("Лучший accuracy score: %.5f" % best_rfc_acc_score)
```

### Gradient Boosting Classifier

```python
gbc = GradientBoostingClassifier(random_state=RANDOM_SEED)
```

```python
gbc_param = {"n_estimators": np.arange(10, 101, 10), "max_depth": np.arange(1, 20), 
          "criterion": ["friedman_mse", "mse"]}
```

```python
# подбор лучших параметров с помощью GridSearchCV
grd_gbc = GridSearchCV(estimator=gbc, param_grid=gbc_param, scoring=metrica, cv=skf, n_jobs=-1)
grd_gbc.fit(X_train, y_train)
```

```python
print('Лучшие параметры:', grd_gbc.best_params_)
print('Лучший accuracy score:', grd_gbc.best_score_)
```

```python
best_gbc = GradientBoostingClassifier(criterion='friedman_mse', max_depth=2, n_estimators=30,
                                     random_state=RANDOM_SEED)
```

```python
best_gbc.fit(X_train, y_train)
y_pred = best_gbc.predict(X_test)
```

```python
gbc_featimp = pd.Series(best_gbc.feature_importances_, X_train.columns).sort_values(ascending=False)[:10]
gbc_featimp
```

*Наблюдается заметное отличие от таблицы важных признаков Random Forest Classifier. Хотя и там, и здесь самыми важными признаками оказались absences и goout, а также школьные оценки были восприняты как значимые, ранжирование и список в целом изменились. Возможно, бустинговая модель учла ошибки прошлых деревьев, что изменило степень значимости признаков по сравнению с тем, как ее оценивал RFC.*

```python
gbc_cvres = pd.DataFrame(grd_gbc.cv_results_)
gbc_cvres.head()
```

```python
gbc_bestres = gbc_cvres.sort_values('mean_test_score', ascending=False)[:5]
gbc_bestres
```

```python
gbc_acc_scores = []

for i in range(5):
    gbc_5 = GradientBoostingClassifier(n_estimators=gbc_bestres.iloc[i]['param_n_estimators'],
                                  max_depth=gbc_bestres.iloc[i]['param_max_depth'],
                                  criterion=gbc_bestres.iloc[i]['param_criterion'],
                                  random_state=RANDOM_SEED)
    gbc_5.fit(X_train, y_train)
    y_pred = gbc_5.predict(X_test)
    gbc_acc_scores.append(accuracy_score(y_pred, y_test))
    print('n_estimators:', gbc_bestres.iloc[i]['param_n_estimators'], ',',
          'max_depth:', gbc_bestres.iloc[i]['param_max_depth'], ',',
          'criterion:', gbc_bestres.iloc[i]['param_criterion'])
    print('accuracy score:', accuracy_score(y_pred, y_test))
    print(40 * '-') 
```

```python
best_gbc_acc_score = max(gbc_acc_scores)
print("Лучший accuracy score: %.5f" % best_gbc_acc_score)
```

### Ada Boost Classifier

```python
ada = AdaBoostClassifier(random_state=RANDOM_SEED)
```

```python
ada_param = {'n_estimators':np.arange(10, 101, 10), 'algorithm':['SAMME', 'SAMME.R']}
```

```python
grd_ada = GridSearchCV(estimator=ada, param_grid=ada_param, scoring=metrica, n_jobs=-1, cv=skf)
grd_ada.fit(X_train, y_train)
```

```python
print('Лучшие параметры:', grd_ada.best_params_)
print('Лучший accuracy score:', grd_ada.best_score_)
```

```python
ada_cvres = pd.DataFrame(grd_ada.cv_results_)
ada_cvres.head()
```

```python
ada_bestres = ada_cvres.sort_values('mean_test_score', ascending=False)[:5]
ada_bestres
```

```python
ada_acc_scores = []

for i in range(5):
    ada_5 = AdaBoostClassifier(n_estimators=ada_bestres.iloc[i]['param_n_estimators'],
                                  algorithm=ada_bestres.iloc[i]['param_algorithm'],
                                  random_state=RANDOM_SEED)
    ada_5.fit(X_train, y_train)
    y_pred = ada_5.predict(X_test)
    print('n_estimators:', ada_bestres.iloc[i]['param_n_estimators'], ',',
          'algorithm:', ada_bestres.iloc[i]['param_algorithm'])
    print('accuracy score:', accuracy_score(y_pred, y_test))
    ada_acc_scores.append(accuracy_score(y_pred, y_test))
    print(40 * '-') 
```

```python
best_ada_acc_score = max(ada_acc_scores)
print("Лучший accuracy score: %.5f" % best_ada_acc_score)
```
